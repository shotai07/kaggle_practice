# Library

```python
import os
import pandas as pd
import numpy as np
from pycaret.regression import *

import matplotlib.pyplot as plt
import seaborn as sns
import sweetviz
from pandas_profiling import ProfileReport
from autoviz.AutoViz_Class import AutoViz_Class
%matplotlib inline
```

# Config

```python
DATA_DIR = '../data/'
INPUT_DIR = DATA_DIR + '01_raw/'
```

```python
ID_COL = 'Id'
Y_COL = 'SalePrice'
```

# Load data

```python
# train
train_fname = INPUT_DIR + 'train.csv'
base_train_df = pd.read_csv(train_fname)
base_train_df.head()
```

```python
# test
test_fname = INPUT_DIR + 'test.csv'
base_test_df = pd.read_csv(test_fname)
base_test_df.head()
```

```python
# submission_example
example_fname = INPUT_DIR + 'sample_submission.csv'
example_df = pd.read_csv(example_fname)
example_df.head()
```

# Baseline model

```python
train_df = base_train_df.drop(ID_COL, axis=1)
train_df.head()
```

```python
s = setup(
    train_df,
    target = Y_COL,
    session_id = 0,
    fold = 5
)
```

```python
best = compare_models()
```

```python
evaluate_model(best)
```

# EDA

## Minimal processing

```python
eda_train_df = base_train_df.copy().assign(flg='train')
eda_test_df = base_test_df.copy().assign(flg='test')
eda_test_df[Y_COL] = None

eda_df = pd.concat([eda_train_df, eda_test_df])
eda_df.head()
```

## Report by Sweetviz

```python
sv_report_name = 'sweetviz_report.html'
if os.path.exists(sv_report_name):
    print('*Already report created.')
else:
    my_report = sweetviz.compare([base_train_df, "Train"], [base_test_df, "Test"], Y_COL)
    #
    my_report.show_html(sv_report_name)
```

# Report by pandas-profiling

```python
pdp_report_name = 'pandas-profiling_report.html'
if os.path.exists(pdp_report_name):
    print('*Already report created.')
else:
    profile = ProfileReport(base_train_df, minimal=True)
    #
    profile.to_file(pdp_report_name)
```

# Report by Autoviz

```python
aviz_report_name = './AutoViz_Plots/'
if os.path.exists(aviz_report_name):
    print('*Already report created.')
else:
    aviz = AutoViz_Class()
    aviz.AutoViz('', 
                 dfte=base_train_df, 
                 depVar=Y_COL,
                 chart_format='html'
                )
```
